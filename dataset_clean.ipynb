```python
import pandas as pd
import numpy as numpy

df=pd.read_csv("Astram event data_anonymized - Astram event data_anonymizedb40ac87.csv");
df.head(1)
```

Column Cleaning -> Discarding unwanted column

```python
#Droping unwanted coloumns
df.drop(['id',
'client_id',
'created_by_id',
'last_modified_by_id',
'assigned_to_police_id',
'citizen_accident_id',
'kgid', 
'closed_by_id',
'resolved_by_id',
'gba_identifier',
'veh_no',
'authenticated',
'map_file',    
'route_path',
'meta_data',
'comment',
'reason_breakdown',
'cargo_material',
'age_of_truck',
'veh_type',
'resolved_at_latitude', 'resolved_at_longitude','resolved_at_address','modified_datetime'], axis=1, inplace=True)

df.head(5)
```

```python
df.columns = df.columns.str.strip()

ordered_columns = [
    'event_type',
    'event_cause',
    'description',
    'requires_road_closure',
    'status',
    'priority',
    'start_datetime',
    'end_datetime',
    'created_date',
    'closed_datetime',
    'resolved_datetime',
    'latitude',
    'longitude',
    'endlatitude',
    'endlongitude',
    'end_address',
    'address',
    'corridor',
    'junction',
    'zone',
    'direction',
    'police_station'
]

# 3. Reorder the DataFrame
df = df[ordered_columns]

df.head(2)
```

```python
#Writing the cleaned data to a new csv file
df.to_csv("cleaned_astram_event_data.csv", index=False) 
```
