```python
import pandas as pd
import numpy as numpy

df=pd.read_csv("Astram event data_anonymized - Astram event data_anonymizedb40ac87.csv");
df.head(1)
```

Column Cleaning -> Discarding unwanted column

```python
#Droping unwanted coloumns
df.drop(['id',
'client_id',
'created_by_id',
'last_modified_by_id',
'assigned_to_police_id',
'citizen_accident_id',
'kgid', 
'closed_by_id',
'resolved_by_id',
'gba_identifier',
'veh_no',
'authenticated',
'map_file',    
'route_path',
'meta_data',
'comment',
'reason_breakdown',
'cargo_material',
'age_of_truck',
'veh_type',
'resolved_at_latitude', 'resolved_at_longitude','resolved_at_address','modified_datetime'], axis=1, inplace=True)

df.head(5)
```

```python
df.columns = df.columns.str.strip()

ordered_columns = [
    'event_type',
    'event_cause',
    'description',
    'requires_road_closure',
    'status',
    'priority',
    'start_datetime',
    'end_datetime',
    'created_date',
    'closed_datetime',
    'resolved_datetime',
    'latitude',
    'longitude',
    'endlatitude',
    'endlongitude',
    'end_address',
    'address',
    'corridor',
    'junction',
    'zone',
    'direction',
    'police_station'
]

# 3. Reorder the DataFrame
df = df[ordered_columns]

df.head(2)
```

Next Task : Understand the Event Dataset

```python
print(df_events['requires_road_closure'].value_counts(dropna=False))

print(df_events['priority'].value_counts(dropna=False))

print(df_master['event_cause'].value_counts(dropna=False))
```

```python
df_events['start_datetime'] = pd.to_datetime(df_events['start_datetime'], errors='coerce', format='mixed').dt.tz_localize(None)
df_events['end_datetime'] = pd.to_datetime(df_events['end_datetime'], errors='coerce', format='mixed').dt.tz_localize(None)

# 3. Recalculate your true error count now
actual_start_errors = df_events.loc[df_events['start_datetime'].isna() & df['start_datetime'].notna()]
actual_end_errors = df_events.loc[df_events['end_datetime'].isna() & df['end_datetime'].notna()]

print(f"Actual text formatting errors (Start): {len(actual_start_errors)}")
print(f"Actual text formatting errors (End): {len(actual_end_errors)}")

# 4. Calculate your duration metric safely
df_events['duration_hours'] = (df_events['end_datetime'] - df_events['start_datetime']) / pd.Timedelta(hours=1)

df_events['duration_hours'].describe()

```

```python
# Count negative durations
negative_durations = df_events[df_events['duration_hours'] < 0]
print(f"Number of rows with negative duration: {len(negative_durations)}")

# Display the problematic rows to investigate
negative_durations[['start_datetime', 'end_datetime', 'duration_hours','event_cause']].head()
```

```python
#check the category of negative value grouped with event cause
negative_summary = negative_durations.groupby('event_cause')['duration_hours'].agg(['count', 'min', 'max'])
print(negative_summary)
```

```python
# Define the thresholds in hours
durations = {
    "24 hours": 24,
    "7 days": 7 * 24,
    "30 days": 30 * 24,
    "365 days": 365 * 24
}

# Calculate and print counts without inspecting rows
for label, hours in durations.items():
    count = (df_events['duration_hours'] > hours).sum()
    print(f"Events longer than {label}: {count}")
```

```python
# Filter for durations greater than 365 days and select specific columns
columns_to_show = ['event_cause', 'start_datetime', 'end_datetime', 'duration_hours', 'status', 'priority']
df_long_events = df_events.loc[df_events['duration_hours'] > (365 * 24), columns_to_show]

# Display the 3 records
df_long_events
```
